```python
import pandas as pd

file_path = '제주특별자치도_신재생에너지발전시설현황_20221116.csv'
df = pd.read_csv("C:\\Users\\gyqls\\OneDrive\\바탕 화면\\제주특별자치도_신재생에너지발전시설현황_20221116.csv", encoding='cp949')
```

```python
df.head()
```

```python
df_sun = df[df['원동력종류']=='태양광']
df_sun.reset_index(drop=True, inplace=True)
df_sun

df_sun['설치장소'] = df_sun['설치장소'].str.replace('제주특별자치도', '')

geo_addr = df_sun['설치장소'].unique()

geo_pos = []

for addr in geo_addr:
    geo_pos.append(addr.replace('시', ''))

print(geo_pos)

from geopy.geocoders import Nominatim

unique_addr = []
unique_lat = []
unique_lng = []
null_geos = []

def geocoding(address):
    geolocator = Nominatim(user_agent='South Korea', timeout=None)
    geo = geolocator.geocode(address)
    if geo is None:
        return None
    else:
        return geo.latitude, geo.longitude

for addr in geo_pos:
    location = geocoding(addr)
    if location is None:
        print('Geo is null:', addr)
    else:
        lat, lng = location
        unique_addr.append(addr)
        unique_lat.append(lat)
        unique_lng.append(lng)
        print(lat, lng)


df_unique_sun = pd.DataFrame({'주소': unique_addr,
                              '경도(x)': unique_lng,
                              '위도(y)': unique_lat})

import pandas as pd
import folium

jeju_map = folium.Map(location=[33.38, 126.55], zoom_start=11)

for i in range(len(df_unique_sun)):
    pos_list = [df_unique_sun.iloc[i]['위도(y)'], df_unique_sun.iloc[i]['경도(x)']]
    address = df_unique_sun.iloc[i]['주소']
    folium.Marker(pos_list, popup=address).add_to(jeju_map)

jeju_map.save('jeju_sun.html')
```

```python
import folium

# 지도 초기화
combined_map = folium.Map(location=[33.38, 126.55], zoom_start=11)

# 태양광 발전 지도 추가
jeju_sun_map = folium.FeatureGroup(name='태양광')
for i in range(len(df_unique_sun)):
    pos_list = [df_unique_sun.iloc[i]['위도(y)'], df_unique_sun.iloc[i]['경도(x)']]
    address = df_unique_sun.iloc[i]['주소']
    folium.Marker(pos_list, popup=address, icon=folium.Icon(color='red')).add_to(jeju_sun_map)
jeju_sun_map.add_to(combined_map)

# 풍력 발전 지도 추가
jeju_wind_map = folium.FeatureGroup(name='풍력')
for i in range(len(df_unique_wind)):
    pos_list = [df_unique_wind.iloc[i]['위도(y)'], df_unique_wind.iloc[i]['경도(x)']]
    address = df_unique_wind.iloc[i]['주소']
    folium.Marker(pos_list, popup=address, icon=folium.Icon(color='blue')).add_to(jeju_wind_map)
jeju_wind_map.add_to(combined_map)

# 레이어 컨트롤 추가
folium.LayerControl().add_to(combined_map)

# 지도 저장
combined_map.save('combined_map.html')
```

```python
# 가입 없이 주소->좌표 변환
from geopy.geocoders import Nominatim

unique_addr=[]
unique_lat=[]
unique_lng=[]

def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    if geo==None:
        print('geo is null')
        return 0, 0
    else:
        return geo.latitude, geo.longitude
```

```python
for addr in geo_pos:
    print(addr) 
    lat, lng = geocoding(addr)
    if lat==0 or lng==0:
        continue
    unique_addr.append(addr)
    unique_lat.append(lat)
    unique_lng.append(lng)
    print(lat, lng)
```
